Donald trump seems to tweet alot of pictures/videos which are url links that at this point we are unable to interpret, hence we decided to remove url links from the text as it does not provide any information. In this inital cleaning we also remove punctuations and the twitter handler sign @

```python
#create function to clean the text
def clean_text(text):
    return " ".join(re.sub(r"(RT\s@[a-zA-Z0-9_]+)|([^A-Za-z0-9])|(amp;)|(\w+://.+)" , " ", text).split()).lower()
```

### Trading days

The US stock market open from Monday to Friday, 930am to 430pm. In order to classify if tweets have impact on the stock market, a label will be assigned to each tweet on how the stock market perform on that day. 1 will be positive and 0 for negative changes in the closing price. As President Trump tweets everyday at different timings, a new feature trade_date will be added to categorize tweets that were posted on that day before market close and any tweets after that will consider as the next trading day. Example, tweet that is posted on 2018-08-10 1:00pm will belong to 2018-08-10(trade_date), if it's posted on 2018-08-10 8:00pm, it will belong to 2018-08-11 (the next day). 

```python
#assign a time for market close (430pm)
mkt_close = datetime.time(16,30)
```

### Topic model

In order to have a better understanding on the context of the tweets, we will be using a technique in NLP called Topic Modelling to assign a related topic to each tweet. But before that will need to tokenize, lemmatize and remove stop words that will affect the modelling.

```python
#initialize spacy
nlp = spacy.load('en_core_web_lg')
```

```python
#create a function to lemmatize
def get_lemmas(text):
    lemmas = []
    doc = nlp(text)
    for token in doc:
        if ((token.is_stop == False) and (token.is_punct == False)) and (token.pos_ != 'PRON'):
            lemmas.append(token.lemma_)
    return lemmas
```

```python
tweets['lemmas_token'] = tweets['clean_text'].apply(lambda t: get_lemmas(t))
```

```python
tweets['lemmas_text'] = tweets['lemmas_token'].apply(lambda w : " ".join(w))
```

```python
tweets.to_csv('../datasets/clean_tweets.csv', index=False)
```

```python
tweets = pd.read_csv('../datasets/clean_tweets.csv')
```

```python
tweets.shape
```

```python
tweets.dropna(inplace=True, subset=['lemmas_text'])
```

```python
tweets.reset_index(inplace=True)
```

```python
#initialize the count vectorizer
cvec = CountVectorizer(stop_words='english')
```

```python
text_data = cvec.fit_transform(tweets['lemmas_text'])
```

```python
text_data.shape
```

```python
# Create and fit the LDA model
lda = LDA(n_components=3, random_state=42)
lda.fit(text_data)
```

```python
#create a function to print out the topics and the key words
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        message = f"Topic #{topic_idx}: "
        message += " ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]])
        print(message)
    print()
```

```python
#print out the topic and key words
print_top_words(lda, cvec.get_feature_names(), 10)
```

```python
#visualize the topic model with pyLDAvis
pyLDAvis.enable_notebook()
pyLDAvis.sklearn.prepare(lda, text_data, cvec)
```

From the keys words and the visualization of the topic model we conclude that President Trump tweets revolved around this 3 topics:

Topic 0: Election Campaign

Topic 1: Immigration

Topic 2: Trade

```python
lda_results = lda.transform(text_data)
```

```python
#columns names
topic_names = [f"topic_{str(i)}" for i in range(lda.n_components)]
```

```python
#add this topic scores to dataframe
tweets[topic_names] = pd.DataFrame(np.round(lda_results, 3))
```

```python
#add column to show dominant topic
tweets['dominant_topic'] = np.argmax(tweets[topic_names].values, axis=1)
```

```python
#give the topic a name instead of number for easier interpretation
tweets['dominant_topic'] = tweets['dominant_topic'].map({0: 'election',
                                                        1: 'immigration',
                                                        2: 'trade'})
```

```python
#create dummy variables for the topics
tweets[['topic_election', 'topic_immigration', 'topic_trade']] = pd.get_dummies(tweets['dominant_topic'], prefix='topic', dtype=int)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.countplot(data=tweets, x='dominant_topic', hue='target', ax=ax)
ax.set_xlabel('Topics')
ax.set_xticklabels(['Trade', 'Election', 'Immigration'], rotation=0)
ax.set_ylabel('Count')
ax.set_title('Tweets Topics', fontsize=15)
```

### Sentiment Analysis

Sentiment Analysis, is a sub-field of Natural Language Processing (NLP) that tries to identify and extract opinions within a given text. The aim of sentiment analysis is to gauge the attitude, sentiments, evaluations, attitudes and emotions of a speaker/writer based on the computational treatment of subjectivity in a text. I will be using VADER (Valence Aware Dictionary and sEntiment Reasoner) to analyze Presdient Trump tweets. ### Part of Speech Tagging & Named entity recognition

```python
#initialize spacy
nlp = spacy.load('en_core_web_sm')
```

```python
# Returns number of verb
def verb_count(text, model=nlp):
    
    # Create doc object
    doc = model(text)
    # Generate list of POS tags
    pos = [token.pos_ for token in doc]
    
    # Return number of proper nouns
    return pos.count('VERB')
```

```python
tweets['verb_count'] = tweets['text'].apply(lambda w : verb_count(w))
```

```python
tweets.verb_count.value_counts()
```

```python
# Returns number of noun
def noun_count(text, model=nlp):
    
    # Create doc object
    doc = model(text)
    # Generate list of POS tags
    pos = [token.pos_ for token in doc]
    
    # Return number of proper nouns
    return pos.count('NOUN')
```

```python
tweets['noun_count'] = tweets['text'].apply(lambda w : noun_count(w))
```

```python
# Returns number of adjective
def adj_count(text, model=nlp):
    
    # Create doc object
    doc = model(text)
    # Generate list of POS tags
    pos = [token.pos_ for token in doc]
    
    # Return number of proper nouns
    return pos.count('ADJ')
```

```python
tweets['adj_count'] = tweets['text'].apply(lambda w : adj_count(w))
```

```python
# Returns number of person
def person_count(text, model=nlp):
    
    # Create doc object
    doc = model(text)
    # Generate list of Ner tags
    ner = [ent.label_ for ent in doc.ents]
    
    # Return number of proper nouns
    return ner.count('PERSON')
```

```python
tweets['person_count'] = tweets['text'].apply(lambda w : person_count(w))
```

```python
# Returns number of org
def org_count(text, model=nlp):
    
    # Create doc object
    doc = model(text)
    # Generate list of Ner tags
    ner = [ent.label_ for ent in doc.ents]
    
    # Return number of proper nouns
    return ner.count('ORG')
```

```python
def score_metrics(model_name, pipe, params,  X_train, y_train, X_test, y_test):
    
    """ 
    Function to fit and collect model metrics
  
    Parameters: 
        tag (str): name for this pipeline
        pipe (pipeline): pipeline to be executed
		param (dict): dictionary of parameters to be used
        X_train (Series): Series holding the train feature
        X_test (Series): Series holding the test feature
        y_train (Series): Series holding the train target
        y_test (Series): Series holding the test target
        
    """
    
    gs = GridSearchCV(pipe, param_grid=params, cv=5, scoring='accuracy')
    gs.fit(X_train, y_train)
    bm = gs.best_estimator_
    y_pred = bm.predict(X_test)
    print(f'{model_name}->Best params: {gs.best_params_}')
    
    metric = {}
    metric['01 Train score'] = f'{bm.score(X_train, y_train):,.4f}'
    metric['02 Test score'] = f'{bm.score(X_test, y_test):,.4f}'
    metric['03 Score diff'] = float(metric['01 Train score'])-float(metric['02 Test score'])

    # calculate Specificity and Sensitivity
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    metric['04 Precision'] = f'{(tp /(tp + fp)):.4f}'
    metric['05 Specificity'] = f'{(tn / (tn + fp)):.4f}'
    metric['06 Sensitivity'] = f'{(tp / (tp + fn)):.4f}'
    metric['07 True Negatives'] = tn
    metric['08 False Positives'] = fp
    metric['09 False Negatives'] = fn
    metric['10 True Positives'] = tp
    metric['11 Train ROC Score'] = f'{roc_auc_score(y_train, bm.predict_proba(X_train)[:,1]):,.4f}'
    metric['12 Test ROC Score'] = f'{roc_auc_score(y_test, bm.predict_proba(X_test)[:,1]):,.4f}'

    # plot roc and confusion matrix
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,5))
    plot_roc_curve(gs, X_test, y_test, name=model_name, ax=ax[0]);
    # Plot baseline. (Perfect overlap between the two populations.)
    ax[0].plot(np.linspace(0, 1, 200), np.linspace(0, 1, 200),
             label='baseline', linestyle='--')
    plot_confusion_matrix(gs, X_test, y_test, ax=ax[1], cmap='RdPu');
    ax[0].set_title(f'ROC for {model_name}');
    ax[1].set_title(f'Confusion Matrix for {model_name}');
    
    return metric
```

```python
def create_embedding_matrix(filepath, word_index, embedding_dim):
    vocab_size = len(word_index) + 1  # Adding again 1 because of reserved 0 index
    embedding_matrix = np.zeros((vocab_size, embedding_dim))

    with open(filepath) as f:
        for line in f:
            word, *vector = line.split()
            if word in word_index:
                idx = word_index[word] 
                embedding_matrix[idx] = np.array(
                    vector, dtype=np.float32)[:embedding_dim]

    return embedding_matrix
```
